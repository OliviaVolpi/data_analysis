The french version of RuPaul's Drag Race premiered in 2022. 
https://www.france.tv/slash/drag-race-france/

Using Google Trends API, we will if, when and how it appeared in Google searches. 


The code in this Notebook is inspired by the great tutorial provided by Aman Kharwal :
https://thecleverprogrammer.com/2021/04/27/google-search-analysis-with-python/





```python
import pandas as pd
from pytrends.request import TrendReq
import matplotlib.pyplot as plt
trends = TrendReq()
```

## Where is "Drag Race France" a popular search term ? 

```python
trends.build_payload(kw_list=["Drag Race France"])
data = trends.interest_by_region()
data = data.sort_values(by="Drag Race France", ascending=False)
data = data.head(10)
data.reset_index().plot(x="geoName", y="Drag Race France", 
                        figsize=(18,13), kind="bar")

plt.xticks(rotation=45)
plt.style.use('ggplot')
plt.show()
```

#### How to read this bar chart? 

__France__, with a score of __100__, is the location where the search term has __the most popularity__. 

Google Trends API considers La RÃ©union as a location of its own, independent of France. 

The next scores show how popular the search terms are by comparison with the location where it is the most popular. A score of 25, observed in Belgium, means "Drag Race France" is searched 4 times less than in France. 


## How did "Drag Race France" popularity as a search term  evolve over time ? 

#### Over the last 12 months 

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Drag Race France'],timeframe='today 12-m',geo='FR')
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(18,13))
data['Drag Race France'].plot()
plt.style.use('ggplot')
plt.title('Total Google Searches for Drag Race France', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

Obviously, "Drag Race France" wasn't a very popular search term in France before __June 2022__, when the first season was first advertised to a __broad audience__. 

The __small peaks__ appearing at the end of 2021 and around March 2022 might be related to the __casting and filming of the show__. 

####  Between June and September

```python
data = TrendReq(hl='en-US', tz=360)
data.build_payload(kw_list=['Drag Race France'],timeframe='2022-06-01 2022-09-30',geo='FR')
data = data.interest_over_time()
fig, ax = plt.subplots(figsize=(18,13))
data['Drag Race France'].plot()
plt.style.use('ggplot')
plt.title('Total Google Searches for Drag Race France', fontweight='bold')
plt.xlabel('Year')
plt.ylabel('Total Count')
plt.show()
```

Popularity for "Drag Race France" as a search term peaked at the __begining and end__ of the show : the last week of June (first episode is broadcasted on France Television, on primetime), and finale, on August 11th. Although the episode was broadcasted on a later time slot, it received attention from mainstream medias, explaining the peak.
